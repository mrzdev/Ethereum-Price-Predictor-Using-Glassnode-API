# Webscraping for the Glassnode Endpoints

Glassnode offers a variety of metrics for on-chain analysis, each with it's own unique endpoint. Here I am webscraping the API homepage to create a list of all of the endpoints available for my API call. 

```python
# Importing necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

The individual endpoints are organized by category, here I am creating a list of each of the category URL's prior to webscraping. 

```python
url_list = [
            "https://docs.glassnode.com/api/addresses",
            "https://docs.glassnode.com/api/blockchain",
            "https://docs.glassnode.com/api/derivatives", "https://docs.glassnode.com/api/distribution",
            "https://docs.glassnode.com/api/entities", "https://docs.glassnode.com/api/fees",
            "https://docs.glassnode.com/api/indicators", "https://docs.glassnode.com/api/institutions",
            "https://docs.glassnode.com/api/market", "https://docs.glassnode.com/api/mining",
            "https://docs.glassnode.com/api/protocols", "https://docs.glassnode.com/api/defi",
            "https://docs.glassnode.com/api/supply", "https://docs.glassnode.com/api/transactions"]
```

Here I am using a list comprehension to webscrape the category URL's specified above and creating a new list that includes all of the endpoints that Glassnode has available. 

```python
urls = []

for url in url_list:
    headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
    page = url
    pageTree = requests.get(page, headers=headers)
    pageSoup = BeautifulSoup(pageTree.content)
    string = "https://api.glassnode.com"
    alpha = pageSoup.find_all('div',{"class":"css-1rynq56 r-gg6oyi r-1b43r93 r-hbpseb r-1nf4jbm r-b88u0q r-1jpmnxg r-1xnzce8"})
    
    for a in list(range(len(alpha))):
        urls.append(string + str(alpha[a]).split('">')[1].split("</div>")[0])         
```

```python
urls
```

```python
urls = pd.DataFrame(urls)
```
